```python
# Сначала дубликат нужной нам части ДЗ1
import nltk
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet as wn
```

```python
exemplar = wn.synset('search.v.01')
print (exemplar.lemma_names())
```

```python
nltk.download('omw-1.4')
nltk.download('omw')
```

```python
# Вывод всех лемм синсета на доступных языках
lst = []
for i in wn.langs():
    lst.append(exemplar.lemma_names(f'{i}'))
lst
```

```python
# Для каждой леммы из каждого языка составляем список синсетов, к которым она относится
synsets_by_lemma = {}
for i in wn.langs():
    for a in lst:
        for b in a:
            synsets = wn.synsets(b, lang=i)
            for synset in synsets:
                if b not in synsets_by_lemma:
                    synsets_by_lemma[b] = [synset]
                else:
                    synsets_by_lemma[b].append(synset)
synsets_by_lemma
```

```python
# Создание списка синсетов, к которым относится не больше 3 лемм из изначального списка
relevant_synsets = []
for lemma in synsets_by_lemma:
    synsets = synsets_by_lemma.get(lemma, [])
    for synset in synsets:
        for eterty in wn.langs():
            if len(synset.lemma_names(eterty)) > 3 and synset not in relevant_synsets:
                relevant_synsets.append(synset)
relevant_synsets
```

```python
words = []
for c in lst:
    for d in c:
        words.append(d)
words
```

```python
# Теперь ДЗ4
# Шаг 1
# Создание пустой матрицы
import pandas as pd
matrix = pd.DataFrame(index=[s.name() for s in relevant_synsets], columns=words)
matrix = matrix.fillna(0)  # Заполнение матрицы нулями

# Заполнение «типологической» матрицы
for relevant_synset in relevant_synsets:
    for word in words:
        for etert in wn.langs():
            if word in relevant_synset.lemma_names(etert):
                matrix.loc[relevant_synset.name(), word] = 1
                break

# Вывод матрицы
print(matrix)
```

```python
array = matrix.values
print (array.shape)
```

```python
from sklearn import manifold
from sklearn.metrics import euclidean_distances

# Шаг 2. Сокращение размерности матрицы до 2 методом MDS
similarities = euclidean_distances(array)
mds = manifold.MDS(n_components=2)
print (mds.fit(similarities).stress_) # функция потерь: показывает, насколько сильно расстояния в новом пространстве отличаются от расстояний в исходном пространстве  
```

```python
array_mds = mds.fit_transform(array)
print (array_mds)
```

```python
x = []
y = []
for pair in array_mds:
    x.append(pair[0])
    y.append(pair[1])
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize = [20, 20])
phrases = matrix.index
for i in range(len(x)):
    plt.scatter(x[i], y[i], c = 'black', s=10)
    plt.annotate(phrases[i], (x[i], y[i]+0.3))
plt.show()
```

```python
mds3d = manifold.MDS(n_components=3, random_state = 42)
print(mds3d.fit(similarities).stress_) # функция потерь: показывает, насколько сильно расстояния в новом пространстве отличаются от расстояний в исходном пространстве  
```

```python
array_mds3d = mds3d.fit_transform(matrix.values)
```

```python
x = []
y = []
z = []
for triple in array_mds3d:
    x.append(triple[0])
    y.append(triple[1])
    z.append(triple[2])
```

```python
fig = plt.figure(figsize=[15,15])
ax = fig.add_subplot(111, projection='3d')
phrases = matrix.index
for i in range(len(x)):
    ax.text(x[i], y[i]+0.3, z[i], phrases[i])
ax.scatter(x, y, z)
plt.show()
```
